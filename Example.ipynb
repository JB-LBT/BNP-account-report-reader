# Processing multiple files

The module offers the possibility to process multiple files together by giving all their names, or even by just providing the folder they're in.

In that case, it will save all the statements in multiple sheets of the given excel file.

## Given the files name

```python
process_files(files=["Data\RLV_CHQ_300040085400001025859_20240722.pdf",
                     "Data\RLV_CHQ_300040085400001025859_20240722.pdf"],
                     dest_file="Output/summary.xlsx")
```

## Given the folder

```python
process_folder(folder="Data", dest_file = "Output/summary_2.xlsx")           # Process all the pdf files in the folder and save the summary in the output folder
```
